```python
import requests
```

```python
import pandas as pd
```

```python
from datetime import datetime

now = datetime.now()
```

```python
url = "https://api.fablabs.io/0/labs.json"
response = requests.get(url)
data = response.json()
```

```python
df = pd.DataFrame(data)
```

```python
df = df.drop(columns=['parent_id', 'blurb', 'avatar_url', 'header_url', 'capabilities', 'links'])
```

```python
df = data.rename(columns={'id': 'fablabs_id'})
```

```python
df = df.rename(columns={'kind_name': 'fablabs_kind_name'})
```

```python
# df = df.rename(columns={'slug': 'fablabs_slug'})
df['fablabs_url'] = 'https://www.fablabs.io/labs/' + df.slug
df = df.drop(columns=['slug'])
```

```python
df['address'] = df.address_1.astype(str) + ', ' + df.address_2
```

```python
df = df.drop(columns=['address_1', 'address_2'])
```

```python
df
```

```python
df = df.rename(columns={'activity_status': 'fablabs_activity_status'})
```

```python
cols = df.columns.tolist()
```
